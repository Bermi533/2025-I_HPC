```python
# plot_delta.py
import numpy as np
import matplotlib.pyplot as plt

# Cargar columnas: N, S1, S2, Delta
data = np.loadtxt("datos.txt", skiprows=1)
N     = data[:,0]
Delta = data[:,3]

plt.figure()
plt.loglog(N, Delta, marker='.', linestyle='none')
plt.xlabel("N")
plt.ylabel(r"$ \left | 1-\frac{S_1}{S_2}  \right | $")
plt.title("Diferencia relativa entre suma directa e inversa en funcin de N")
plt.grid(True, which="both", ls=":")
plt.tight_layout()
plt.savefig("delta_vs_N.pdf")
```
