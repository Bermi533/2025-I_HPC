```python
#!/usr/bin/env python3
import numpy as np
import matplotlib.pyplot as plt

# List of your files and corresponding labels
files  = ['seed1.txt', 'seed2.txt', 'seed5.txt']
labels = ['seed = 1',    'seed = 2',    'seed = 5']

plt.figure(figsize=(8,5))
for fname, label in zip(files, labels):
    # Each file has two columns: x_center and pdf_estimate
    data = np.loadtxt(fname)
    x, pdf = data[:,0], data[:,1]
    plt.plot(x, pdf, 'o', linestyle = '-', label=label, markersize = 4)

plt.title('Distribuciones de densidad de probabilidad (SEEDS =1,2,5)')
plt.xlabel('x')
plt.ylabel('Distribución')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig('three_pdfs.pdf') 
plt.show()
```
